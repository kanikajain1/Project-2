## Connect to local Database

```python
rds_connection_string = "kanika:postgres@localhost:5432/crimes_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for local tables

```python
engine.table_names()
```

### Use pandas to load merged DataFrame into database

```python
final_crimes_df.to_sql("crimes_table", con=engine, if_exists='append', index=False)
```
